```python
#conditional column : case ,ending  case is necessary
q=""" select i.customer_id,c.first_name || " " || c.last_name Name,sum(i.total) total, count(i.invoice_id) id,
    case 
    when sum(i.total)<40 then "small spender"
    when sum(i.total)>100 then "big spender"
    else "regular"
    end customer_category
    from customer c
    inner join invoice i
    on
    i.customer_id= c.customer_id
    group by c.customer_id
    order by 3 desc

""" 
pd.read_sql_query(q, conn)
```

```python
#track id ,no of playlist track and playlist name,total duration in seconds

#can use avg(for columns) in having
```

```python
q=""" select t.track_id,t.name Track_name,p.playlist_id,p.name "play_list name",
    t.milliseconds seconds
    from playlist_track pt
    inner join track t
    on
    pt.track_id=t.track_id
    inner join playlist p
    on 
    p.playlist_id=pt.playlist_id
    group by 1
    having avg(t.name)>100 
""" 
pd.read_sql_query(q, conn)
```
