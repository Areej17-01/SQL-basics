```python

q="""select
    count(major_category) as "total count"
    FROM
     recent_grads
     where 
     ShareWomen>0.5 and
     Major_category=="Engineering"
     limit 5"""
pd.read_sql_query(q,conn)
```

```python
#data aggregation returns only one row,following that aplying limit func has no value and use,count is a aggregate func
q="""select
    count(major_category) as "total count",count(major_category) as total_count
    FROM
     recent_grads
     where 
     ShareWomen>0.5 and
     Major_category=="Engineering"
     """
pd.read_sql_query(q,conn)
```

```python

q="""select
    distinct(major_category) 
    FROM
     recent_grads
     where 
     ShareWomen>0.5 and
     Major_category=="Engineering"
     """
pd.read_sql_query(q,conn)
```

```python
#data aggregation returns only one row,following that aplying limit func has no value and use,count is a aggregate func
q="""select
    major_category,avg(sharewomen) as  "sharewomen avg"
    FROM
     recent_grads
    group by major_category
     order by "sharewomen avg" DESC
     limit 3
     """
pd.read_sql_query(q,conn)
```

```python
##cant apply where
q="""select
    major_category,avg(sharewomen) as  "sharewomen avg"
    FROM
     recent_grads
    group by major_category
    
    having
    "sharewomen avg" >0.5
     
     """
pd.read_sql_query(q,conn)
```
